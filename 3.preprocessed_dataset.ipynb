```python
import re

# Define the pattern for removing suffixes
pattern = r'\b(' + '|'.join(re.escape(suffix) for suffix in removal_list) + r')\b'

# Combine all cleaning steps, modifying the problematic line
mapped_df.loc[:, 'NAME_CLEANED'] = (
    mapped_df['NAME']
    .str.lower()  # Convert to lowercase
    .str.replace(r'[^\x00-\x7F]', '', regex=True)  # Replace non-ASCII characters with space
    .str.replace(r'[^a-zA-Z0-9\s]', '', regex=True)  # Replace special characters
    .str.replace(pattern, '', regex=True)  # Replace suffixes
    .str.replace('[.,]', '', regex=True)  # Replace commas and periods
    .str.strip()  # Remove leading and trailing spaces
)

# remove non-ASCII rows
mapped_df = mapped_df[mapped_df['NAME_CLEANED'] != '']
```
