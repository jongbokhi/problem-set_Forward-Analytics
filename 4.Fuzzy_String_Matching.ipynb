```python
import pandas as pd
from fuzzywuzzy import fuzz
```

```python
full_df = pd.read_csv('/content/drive/MyDrive/forward_analytics_results/companydataset_peopledata.csv')
```

```python
df = full_df.drop('SIZE', axis=1)
```

```python
smaple_df = df.sample(n=5000, random_state=42)
```

```python
threshold = 90


def find_duplicates(df, column_name, method, threshold):
    """
    Finds duplicate rows in a DataFrame based on string similarity.

    Args:
    - df (pd.DataFrame): The input DataFrame.
    - column_name (str): The column to check for duplicates.
    - method (str): The fuzz method to use ('ratio', 'partial_ratio', 'token_sort_ratio', etc.).
    - threshold (int): The similarity threshold (0-100).

    Returns:
    - List of tuples with (name1, name2, similarity).
    """
    # Validate method
    if not hasattr(fuzz, method):
        raise ValueError(f"Invalid method '{method}' provided. Available methods: ratio, partial_ratio, token_sort_ratio, etc.")

    # Resolve the fuzz method dynamically
    fuzz_method = getattr(fuzz, method)

    duplicates = []
    names = df[column_name].tolist()

    # Compare names
    for i in range(len(names)):
        for j in range(i + 1, len(names)):
            similarity = fuzz_method(names[i], names[j])
            if similarity >= threshold:
                duplicates.append((names[i], names[j], similarity))

    return duplicates
```

```python
token_duplicates = find_duplicates(smaple_df, 'NAME', 'token_sort_ratio', threshold)
```

```python
token_duplicates_df = pd.DataFrame(token_duplicates, columns=['NAME1', 'NAME2', 'Similarity'])
```
