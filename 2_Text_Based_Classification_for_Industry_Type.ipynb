```python
import pandas as pd

from itertools import chain
import nltk
nltk.download('punkt_tab')
```

```python
df = pd.read_excel("/content/drive/MyDrive/forward_firm_universe.xlsx")
```

```python
# Lowercase strings in a cloumn
df['Entity_Name'] = df['Entity_Name'].str.lower()
```

```python
# processes a text column by removing any non-alphabetic characters,
# then breaks the text into individual words (tokenizes it),
# and generates a new DataFrame where each row contains a single word.

names = df['Entity_Name'].str.replace('[^A-z]', ' ', regex=True).str.replace(' +', ' ',  regex=True).str.strip()

split_names=[nltk.word_tokenize(str(name)) for name in names]

split_names=list(map(str, chain.from_iterable(split_names)))

names_df=pd.DataFrame({'word':split_names})
```

```python
names_df.head(50)
```

```python
# Counting how often each word appears in the names

count_df = pd.DataFrame(names_df.groupby('word')['word'].count().sort_values(ascending=False).reset_index(name='counts'))
```

```python
# Create a list of common words that indicate industries.
industry_category = count_df['word'].to_list()[:101]
print(industry_category)
```

```python
## Classify industries using a list of common words that indicate industries

%%time
df['Industry_Classification'] = None

for i in industry_category:
  # Use .any() to check if any value in the Series is True
  if df['Entity_Name'].str.contains(i, case=False).any():
    # Use .loc to assign the value to the rows where the condition is True
    df.loc[df['Entity_Name'].str.contains(i, case=False), 'Industry_Classification'] = i
  else:
    None
```
